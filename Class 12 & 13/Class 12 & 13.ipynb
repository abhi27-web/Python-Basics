# Paste Method

```python
from PIL import Image, ImageEnhance

new = Image.new("RGBA", (200,200))[]

img = Image.open("PM.jpg")
img = img.resize((100,100))
new.paste(img, (0,0))
new.paste(img, (100,100))
display(new)
```

```python
new = Image.new("RGBA", (400,400))

img = Image.open("PM.jpg")
img = img.resize((100,100))

new.paste(img, (0,0))
new.paste(img, (200,0))
new.paste(img, (100,100))
new.paste(img, (100,300))
new.paste(img, (300,100))
new.paste(img, (0,200))
new.paste(img, (200,200))
new.paste(img, (300,300))

display(new)
```

## Brightness

```python
collage = Image.new("RGBA", (400,200))

img = Image.open("PM.jpg")
img = img.resize((100,100))

filter = ImageEnhance.Brightness(img)

c=0.7
for i in range(0,500,100):
    for j in range(0,400,100):
        new_img = filter.enhance(c)
        collage.paste(new_img, (j,i))
        c+=0.1
display(collage)
```
