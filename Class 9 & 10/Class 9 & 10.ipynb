```python
# Labels
X = np.random.uniform(50,60,10)
Y = np.arange(10)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(X,Y,color='C2', marker='o')

plt.title("Weight of students", fontdict = font1)
plt.xlabel("X (in kgs)", fontdict = font2)
plt.ylabel("Y (no of students)", fontdict = font2)
```

```python
plt.title("Weight of students", fontdict = font1)
plt.xlabel("X (in kgs)", fontdict = font2)
plt.ylabel("Y (no of students)", fontdict = font2)
plt.plot(X,Y,color='C2', marker='o')
plt.grid()
plt.show()
```

```python
plt.title("Weight of students", fontdict = font1)
plt.xlabel("X (in kgs)", fontdict = font2)
plt.ylabel("Y (no of students)", fontdict = font2)
plt.plot(X,Y,color='C2', marker='o')
plt.grid(axis='y')
plt.show()
```

```python
plt.title("Weight of students", fontdict = font1)
plt.xlabel("X (in kgs)", fontdict = font2)
plt.ylabel("Y (no of students)", fontdict = font2)
plt.plot(X,Y,color='C2', marker='o')
plt.grid(axis='x')
plt.show()
```

```python
plt.title("Weight of students", fontdict = font1)
plt.xlabel("X (in kgs)", fontdict = font2)
plt.ylabel("Y (no of students)", fontdict = font2)
plt.plot(X,Y,color='C2', marker='o')
plt.grid(color = 'green', linestyle = 'dotted', linewidth = 0.5)
plt.show()
```

### Colors

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = '#803030')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#4fe0dc')

plt.show()
```

### Color Each Dot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

### ColorMap

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

# BAR PLOT

```python
X = np.arange(10)
Y = np.random.uniform(0,1,10)

fig, ax = plt.subplots()
ax.bar(X,Y)
```

### Horizontal Bar Plot

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

### Coloured Bars

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red")
plt.show()
```

### Bar Width

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.5)
plt.show()
```

### Bar Height

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```

# PIE CHART

```python
Z = np.random.uniform(0,1,8)
plt.pie(y)
plt.show() 
```

### Labels

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

### Start Angle

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 180)
plt.show() 
```

### Explode

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1,0,0,0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

### Shadow

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

### Colors

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["red", "yellow", "pink", "brown"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

### Legend

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["red", "yellow", "pink", "brown"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.legend()
plt.show() 
```

### Legend With Header

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["red", "yellow", "pink", "brown"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.legend(title="Four Fruits:")
plt.show() 
```
