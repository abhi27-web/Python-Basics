```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
image1=cv2.imread('top_right.jpg')
image2=cv2.imread('bottom_right.jpg')
image3=cv2.imread('top_left.jpg')
image4=cv2.imread('bottom_left.jpg')
image5=cv2.imread('center.jpeg')
```

```python
plt.imshow(image1)
```

```python
rgbim1=cv2.cvtColor(image1,cv2.COLOR_BGR2RGB)
rgbim2=cv2.cvtColor(image2,cv2.COLOR_BGR2RGB)
rgbim3=cv2.cvtColor(image3,cv2.COLOR_BGR2RGB)
rgbim4=cv2.cvtColor(image4,cv2.COLOR_BGR2RGB)
rgbim5=cv2.cvtColor(image5,cv2.COLOR_BGR2RGB)
```

# Understanding Borders

```python
sample=cv2.imread('top_right.jpg')
```

```python
sample=cv2.resize(sample,(200,200))
sample=cv2.copyMakeBorder(sample,0,10,10,0,0)#top-right
plt.imshow(sample)
```
