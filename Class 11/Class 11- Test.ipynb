# Q3

![alt text](Class11-T/3.png "Title")

```python
import numpy as np
Subject = ['English','Hindi','Maths' ,'Science','S.Science']
First_Term = [67,72,88,81,73]
Second_Term = [70,65,95,85,75]
  
X_axis = np.arange(len(Subject))
  
plt.bar(X_axis - 0.2, First_Term , 0.4, label = '1st Term')
plt.bar(X_axis + 0.2, Second_Term , 0.4, label = '2nd Term')
  
plt.xticks(X_axis, Subject)
plt.xlabel("Groups")
plt.ylabel("Number of Students")
plt.title("Number of Students studying each subject")
plt.grid(linewidth = 2)
plt.legend()
plt.show()
```
